```python
import cv2
```

```python
import numpy as np
```

```python
img=cv2.imread("Img/fruits.jpeg")
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
cv2.imshow("original",img_rgb)
```

```python
cv2.waitKey(0)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("Gray Image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_blue=img[:,:,0]
img_green=img[:,:,1]
img_red=img[:,:,2]
```

```python
img_new=np.hstack((img_blue,img_green,img_red))
```

```python
cv2.imshow("original",img_new)
```

```python
cv2.waitKey(0)
```
