```python
import cv2
import numpy as np
```

```python
img=np.zeros((512,512,3))
```

```python
cv2.imshow("original",img)
cv2.waitKey(0)
```

```python
 rectangle=cv2.rectangle(img,pt1=(50,50),pt2=(150,150),color=(0,255,255),thickness=1)
```

```python
cv2.imshow("original",rectangle)
cv2.waitKey(0)
```

```python
circle=cv2.circle(
    img,                # Image jis par draw karna hai
    center=(200, 200),  # Circle ka center (x, y)
    radius=50,          # Radius in pixels
    color=(0,0,255),    # BGR format (Red)
    thickness=-2         # Border thickness
)
```

```python
cv2.imshow("original",circle)
cv2.waitKey(0)
```

```python
L=cv2.line(
    img,
    pt1=(0, 0),
    pt2=(500, 500),
    color=(0, 0, 255),   # Red (BGR)
    thickness=2
)
```

```python
cv2.imshow("original",L)
cv2.waitKey(0)
```
