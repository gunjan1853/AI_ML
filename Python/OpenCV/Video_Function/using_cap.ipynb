```python
import cv2
import numpy as np
```

```python
cap=cv2.VideoCapture(0)
```

```python
while True:
    ret,frame=cap.read()
    if not ret:
        print("Oopps frame not found !")
        break;
    cv2.imshow("video",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
