```python
import cv2
import numpy as np
```

```python
cap=cv2.VideoCapture(0)
```

```python
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
```

```python
compression = cv2.VideoWriter_fourcc(*'XVID')
```

```python
recorder = cv2.VideoWriter('output.avi', compression, 20, (width, height))
```

```python
while True:
    success, frame = cap.read()

    if not success:
        break

    recorder.write(frame)

    cv2.imshow("Recording", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
recorder.release()
cap.release()
cv2.destroyAllWindows()
```
